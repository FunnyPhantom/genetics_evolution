```python
import pygad
import numpy as np
import math
```

```python
# ans mapper
ans = "که عشق اسان نمود اول ولی افتاد مشکل ها"

alphabet= list(" ابپتثجچحخدذرزژسشصضطظعغفقکگلمنوهی")
alph_dict= {k:v for v,k in enumerate(alphabet)} #  index_dict
alphabet_len = len(alphabet)

def default_fitness_func(solution, solution_idx):
    global ans
    c = 0
    for i in range(len(ans)):
        if solution[i] == alph_dict[ans[i]]:
            c += 1
    return c

def ans_to_text(ans):
    text = ""
    for i in ans:
        text += alphabet[int(i)]
    return text

def world(num_generations = 100,num_parents_mating = 2,keep_parents = 1,mutation_percent_genes = 10, population_size = 100, fitness_func=default_fitness_func, crossover_type="two_points", mutation_type="random", parent_selection_type="sss"):
    gene_space = [np.arange(alphabet_len) for i in range(len(ans))]
    initial_solution = [np.array([np.random.choice(x) for x in gene_space]) for i in range(population_size)]
    ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       fitness_func=fitness_func,
                       gene_space=gene_space,
                       random_mutation_min_val=0,
                       random_mutation_max_val=alphabet_len - 1,
                       keep_parents=keep_parents,
                       parent_selection_type=parent_selection_type,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes,
                       initial_population=initial_solution,
                       )
    ga_instance.run()

    # parameters
    print("num parents mating: ",num_parents_mating)
    print("keep parents: ",keep_parents)
    print("num generations: ",num_generations)
    print("population size: ",population_size)
    print("mutation percent genes: ",mutation_percent_genes)
    print("parent selection type: ", parent_selection_type)
    print("crossover type: ", crossover_type)
    print("mutation type: ", mutation_type)
    
    
    ga_instance.plot_fitness()
    
    sol = ga_instance.best_solution()[0]
    
    print(ans_to_text(sol))
    print("ans fitness: ", ga_instance.best_solution()[1])
    

```

```python
world()
```

```python
world(num_parents_mating = 10)
```

```python
world(num_parents_mating = 10,keep_parents = 5)
```

```python
world(num_parents_mating = 10,keep_parents = 5, mutation_percent_genes = 100)
```

```python
world(crossover_type="uniform")
```

```python
for mutation_perc in range(1, 15):
    world(mutation_percent_genes = mutation_perc, crossover_type="uniform")
```

#### We see that with any increase in mutation more than 8 percent, the population varies greatly in terms of their ability to reach max fitness.
