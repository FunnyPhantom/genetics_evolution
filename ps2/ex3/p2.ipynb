The suggested parameter inside the class makes it a bit hard to decipher any meaningful about the population.
So lets play with the parameters.

```r
INIT_POP = 100
simulation = function(offspring_count, living_chance, generations) {
    age = vector("numeric", generations)
    age[1] = INIT_POP
    
    for (j in 1:generations) {
        for (i in (length(age) - 1):1) {
            age[i+1] = age[i] * living_chance
        }
        age[1] = 0;
        age[1] = sum(age) * offspring_count
    }

    print(paste("offspring count: ", offspring_count))
    print(paste("chance: ", living_chance))
    print(paste("num of generations", generations))


    # barplot of the population with x axis age and y axis population
    plot(age, main="Number of population based on their age", xlab="age", ylab="population", col=rainbow(100), type="l")
    plot(log(age + 1), main="Number of population based on their age (log)", xlab="age", ylab="population (log)")

}
```

```r
simulation(offspring_count=3, living_chance=0.9, generations=100)
```
